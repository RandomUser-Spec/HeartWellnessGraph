# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
#read the csv
df = pd.read_csv('framingham.csv')
df
```

```python
#Filter and make a new dataframe for non smokers
non_smokers = df.loc[df['currentSmoker'] == 0]
non_smokers
```

```python
#Group by age column and find average heart rate at different age
avg_heart_rate_age = non_smokers.groupby('age')['heartRate'].mean().reset_index()
avg_heart_rate_age
```

```python
#Filter and make a new dataframe for smokers
smokers = df.loc[df['currentSmoker'] == 1]
smokers
```

```python
#Group by age column and find average heart rate at different age
avg_heart_rate_age_smokers = smokers.groupby('age')['heartRate'].mean().reset_index()
avg_heart_rate_age_smokers
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
plt.figure(figsize = (19,8))
plt.plot(avg_heart_rate_age['age'], avg_heart_rate_age['heartRate'], label = "Non Smoker", linewidth = 3)
plt.plot(avg_heart_rate_age_smokers['age'], avg_heart_rate_age_smokers['heartRate'], label = "Smoker", linewidth = 3)
plt.xlabel('Age')
plt.xticks(rotation = 'vertical')
plt.ylabel('Heart Rate')
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv
pd.read_csv('framingham.csv')
#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
coronary_heart_disease = df.loc[df['TenYearCHD'] == 1]
coronary_heart_disease
```

```python
#Group by age column and count the rows of TenYearCHD column
grouped_heart_disease = coronary_heart_disease.groupby('age')['TenYearCHD'].count().reset_index()
grouped_heart_disease
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
plt.figure(figsize = (19,8))
plt.plot(grouped_heart_disease['age'], grouped_heart_disease['TenYearCHD'], label = "Chances of CHD", linewidth = 3)
plt.xlabel('Age')
plt.xticks(rotation = 'vertical')
plt.ylabel('Chances')
plt.show()
```

Conslusion - 

```python
# bubble graph

plt.figure(figsize = (20,10))
plt.scatter(grouped_heart_disease['age'], grouped_heart_disease['TenYearCHD'], label = 'Chances of Coronary Heart Disease', color = 'yellow', s = grouped_heart_disease['TenYearCHD'] * 10)
plt.xlabel('Age')
plt.xticks(rotation = 'vertical')
plt.ylabel('Chances of Disease')
plt.show()
```
